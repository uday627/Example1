```python
dataset = dataset.drop(columns='Car_Name')
```

```python
dataset['Current Year'] = 2023
```

```python
dataset.head()
```

```python
dataset['Years Old'] = dataset['Current Year'] - dataset['Year']
```

```python
dataset.head()
```

```python
dataset = dataset.drop(columns=['Current Year', 'Year'])
```

```python
dataset.head()
```

```python
dataset.select_dtypes(include='object').columns
```

```python
dataset = pd.get_dummies(data=dataset, drop_first=True)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
x = dataset.drop(columns='Selling_Price')
```

```python
y = dataset['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, Binarizer

classifier_rf = RandomForestClassifier(random_state=42, n_jobs=-1, max_depth=5,
                                       n_estimators=100, oob_score=True)

# Check if y_train is a pandas DataFrame or Series
if isinstance(y_train, pd.DataFrame) or isinstance(y_train, pd.Series):
    print(y_train.head())
    y_train = y_train.values  # Convert DataFrame or Series to numpy array

# Check if y_train is a 1D numpy array
if len(y_train.shape) == 1:
    y_train = y_train.reshape(-1, 1)

# Use Binarizer to transform y_train if necessary
binarizer = Binarizer()
y_train_cat = binarizer.fit_transform(y_train)

# Use LabelEncoder to transform y_train if necessary
label_encoder = LabelEncoder()
y_train_cat = label_encoder.fit_transform(y_train_cat)

# Now you can fit your RandomForestClassifier using x_train and y_train_cat
classifier_rf.fit(x_train, y_train_cat)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Define and train your regressor model
regressor = RandomForestRegressor(random_state=42)
regressor.fit(x_train, y_train)

# Make predictions on the test set
y_pred = regressor.predict(x_test)


y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {
    'n_estimators':[100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'criterion':['absolute_error', 'squared_error'],
    'max_depth':[10, 20, 30, 40, 50],
    'min_samples_split':[2, 5, 10, 20, 50],
    'min_samples_leaf':[1, 2, 5, 10],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
parameters
```

```python
random_cv = RandomizedSearchCV(estimator=regressor, param_distributions=parameters, n_iter=10,
                               scoring='neg_mean_absolute_error', cv=5, verbose=2, n_jobs=-1)
```

```python
print(random_cv.param_distributions['criterion'])
random_cv.fit(x_train, y_train)
```
